# Observation-2

1. From the above corr and sns graph of Global_active_power & Global_intensity  we can conclude that both are very highly correlated (almost 1), so we can drop any one feature from the  above 2- so we have decided to remove global _intensity feature

As one of the easiest way to reduce the dimensionality of a dataset is to remove the highly correlated features. 
The idea is that if two features are highly correlated then the information they contain is very similar, 
and it is likely redundant to include both the features. So it is better to remove one of them from the feature set
